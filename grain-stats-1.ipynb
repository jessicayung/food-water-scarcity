## 1. World (except US) coarse grains: Supply and disappearance (million metric tons) 1960/1-2016/7*

\*2016/7 data is predicted data.

Cleaned data in Google Sheets.

```python
data_path = "FGYearbookTable02_Foreign_Coarse_Grains_Supply_and_Disappearance - FGYearbookTable02.csv"
foreign_grains_data = pd.read_csv(data_path, skiprows = 1, index_col = 2)
foreign_grains_data
```

```python
corn_data = foreign_grains_data.iloc[0:57]
corn_data
```

```python
# Remove first two columns
corn_data = corn_data.drop('Commodity', 1)
corn_data = corn_data.drop('Market Year', 1)
```

```python
corn_data
```

```python
# Clean rest of data
foreign_grains_data = foreign_grains_data.drop('Commodity', 1)
foreign_grains_data = foreign_grains_data.drop('Market Year', 1)
foreign_grains_data
```

```python
# Create sub-dataframes for other grains
# Indices found by observation from spreadsheet
sorghum_data = foreign_grains_data.iloc[59:115]
barley_data = foreign_grains_data.iloc[117:173]
coarse_grains_data = foreign_grains_data.iloc[175:231]
```

```python
plt.plot(corn_data['Ending Stocks'])
plt.title('Corn Ending Stocks for World excluding US')
plt.ylabel('Millions of metric tons')
```

```python
plt.plot(corn_data['Production'])
plt.plot(corn_data['Food, alcohol, and industrial use'])
plt.plot(corn_data['Feed use'])
plt.title('Corn Data, World excluding US')
plt.ylabel('Millions of metric tons')
plt.legend()
```

```python
def prod_two_uses_plot(data, data_name):
    plt.plot(data['Production'])
    plt.plot(data['Food, alcohol, and industrial use'])
    plt.plot(data['Feed use'])
    plt.plot(data['Ending Stocks'])
    plt.title(data_name + ', World excluding US')
    plt.ylabel('Millions of metric tons')
    plt.legend()
prod_two_uses_plot(corn_data, 'Corn Data')
```

```python
prod_two_uses_plot(sorghum_data, 'Sorghum Data')
```

Sorghum Data is much more volatile than the Corn Data and the increasing trend is less significant. Quantity is much smaller. Increasing trend (if any) in Food etc use is less significant. No clear increasing trend from 1980 in Feed Use.

```python
prod_two_uses_plot(barley_data, 'Barley Data')
```

```python
prod_two_uses_plot(coarse_grains_data, 'Coarse Grains Data')
```
